```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, lower, trim

spark = SparkSession.builder \
    .appName("WeatherAnalysis") \
    .getOrCreate()
```

```python
df = spark.read.csv(
    "Weather Training Data.csv",
    header=True,
    inferSchema=True
)

df.show(5)
```

```python
numeric_cols = ["MinTemp", "MaxTemp", "Rainfall", "Evaporation", "Sunshine",
                "WindGustSpeed", "WindSpeed9am", "WindSpeed3pm",
                "Humidity9am", "Humidity3pm", "Pressure9am", "Pressure3pm",
                "Cloud9am", "Cloud3pm", "Temp9am", "Temp3pm"]

for c in numeric_cols:
    df = df.withColumn(c, col(c).cast("double"))

df = df.withColumn(
    "RainToday",
    when(lower(trim(col("RainToday"))).isin("yes", "1"), True).otherwise(False))

df = df.withColumn(
    "RainTomorrow",
    when(lower(trim(col("RainTomorrow"))).isin("yes", "1"), True).otherwise(False)
)
```

```python
df.describe().show()
df.groupBy("Location").avg("MaxTemp").show()
df.filter(col("MaxTemp") > 30).show()
```

General Temperature Distribution

```python
import matplotlib.pyplot as plt
import numpy as np

pdf = df.sample(fraction=0.1).toPandas()

pdf["MaxTemp"].hist(bins=30)
plt.title("Temperature Distribution")
plt.show()

avg_temp_df = df.groupBy("Location").avg("MaxTemp")
```

Average Max Temp by Location

```python
pdf = avg_temp_df.toPandas()

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.bar(pdf["Location"], pdf["avg(MaxTemp)"])
plt.xticks(rotation=90)
plt.title("Average Max Temperature by Location")
plt.xlabel("Location")
plt.ylabel("Average Max Temperature")
plt.tight_layout()
plt.show()
```

Temperature Distribution by Location

```python
locations = df.select("Location").distinct().limit(5).toPandas()["Location"].tolist()

for loc in locations:
    subset = df.filter(col("Location") == loc)

    pdf = subset.select("MaxTemp").dropna().sample(fraction=0.2).toPandas()

    plt.figure()
    plt.hist(pdf["MaxTemp"], bins=30)
    plt.title(f"Temperature Distribution for {loc}")
    plt.xlabel("MaxTemp")
    plt.ylabel("Frequency")
    plt.show()
```

Rainfall vs. Temperature

```python
pdf = df.select("MaxTemp", "Rainfall").dropna().sample(fraction=0.1).toPandas()

x = pdf["MaxTemp"]
y = pdf["Rainfall"]

# Adding Regression line
m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b, color="red")   # <-- Here
plt.title("Rainfall vs Max Temperature")
plt.xlabel("MaxTemp")
plt.ylabel("Rainfall")
plt.show()
```

Wind Speed vs. Temperature Change

```python
from pyspark.sql.functions import col

pdf = df.select("TempChange", "WindGustSpeed").dropna().sample(fraction=0.1).toPandas()

x = pdf["TempChange"]
y = pdf["WindGustSpeed"]

m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b, color="red")
plt.title("Wind Gust Speed vs Temperature Change")
plt.xlabel("Temp Change")
plt.ylabel("Wind Gust Speed")
plt.show()
```

Evaporation vs. Sunshine

```python
pdf = df.select("Sunshine", "Evaporation").dropna().sample(fraction=0.1).toPandas()

x = pdf["Sunshine"]
y = pdf["Evaporation"]

m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b, color="red")
plt.title("Evaporation vs Sunshine")
plt.xlabel("Sunshine")
plt.ylabel("Evaporation")
plt.show()
```
